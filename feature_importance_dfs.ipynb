```python
import psycopg2
from psycopg2 import sql
import pandas as pd
import pandasql as ps
```

```python
# Database connection parameters
db_params = {
    'dbname': 'project4',
    'user': 'postgres',
    'password': 'postgres',
    'host': 'localhost',
    'port': '5432'
}

# Establish the connection
conn = psycopg2.connect(**db_params)
```

```python
query = """
SELECT cognitive_test_scores,
    CASE
        WHEN CAST(Dementia AS INTEGER) = 1 THEN 'Yes'
        ELSE 'No'
    END AS dementia_status,
    COUNT(*)
FROM dementia
GROUP BY cognitive_test_scores, dementia_status
ORDER BY cognitive_test_scores DESC;
"""

query_df = pd.read_sql_query(query, conn)
query_df
```

```python
query2 = """
SELECT
    CASE
        WHEN CAST(apoe_4 AS INTEGER) = 1 THEN 'Positive'
        ELSE 'Negative'
    END AS apoe_4,
    CASE
        WHEN CAST(Dementia AS INTEGER) = 1 THEN 'Yes'
        ELSE 'No'
    END AS dementia_status,
    COUNT(*)
FROM dementia
GROUP BY apoe_4, dementia_status
ORDER BY apoe_4 DESC;
"""
query_df2 = pd.read_sql_query(query2, conn)
query_df2
```

```python
query3 = """
SELECT
    CASE
        WHEN CAST(depression_status AS INTEGER) = 1 THEN 'Yes'
        ELSE 'No'
    END AS depression_status,
    CASE
        WHEN CAST(Dementia AS INTEGER) = 1 THEN 'Yes'
        ELSE 'No'
    END AS dementia_status,
    COUNT(*)
FROM dementia
GROUP BY
    depression_status,
    dementia_status
ORDER BY
    depression_status DESC;
    """
query_df3 = pd.read_sql_query(query3, conn)
query_df3
```
